```python
import pandas as pd
import numpy as np
```

```python
leagues = ['9394', '9495', '9596', '9697', '9798', '9899', '9900', '0001', '0102', '0203', '0304', '0405', '0506', '0607', '0708', '0809', '0910', '1011', '1112', '1213', '1314', '1415', '1516', '1617', '1718', '1819', '1920', '2021', '2122', '2223', '2324', '2425']
```

```python
master_df = pd.DataFrame()

for league in leagues:
    league_df = pd.read_csv(f"data/season-{league}.csv")
    master_df = pd.concat([master_df, league_df], axis=0)

print(master_df)
```

```python
master_df['Date_ddmmyy'] = pd.to_datetime(master_df['Date_ddmmyy'], format='%d/%m/%y')
master_df
```

```python
teams = master_df['H_Team'].unique().tolist()
print(teams, len(teams))
```

```python
grouped = master_df.groupby('H_Team')
H_splits_dfs = {category: group for category, group in grouped}
grouped = master_df.groupby('A_Team')
A_splits_dfs = {category: group for category, group in grouped}

for team in teams:
    home_df = H_splits_dfs[team]
    away_df = A_splits_dfs[team]

    combined_df = pd.concat([home_df, away_df], axis=0)
    sorted_combined_df = combined_df.sort_values(by='Date_ddmmyy')
    sorted_combined_df.to_csv(f'{team}.csv', index=False)
    print(team)
```
